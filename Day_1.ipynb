Volume and Surface Area Calculator

```python
print("Hi, I am Volume and Surface Area Calculator.")
print()
print("What is your name, User?")
name = input()
print("Hi",name)
print()
length=int(input("Give me a length of a cuboid in inches."))
width=int(input("Give me a width of that same cuboid in inches."))
height=int(input("Give me a height of that same cuboid in inches."))
print("The volume of the cuboid is",length*width*height,"inches squared.")
print("The surface area of the cuboid is",2*length*width+2*length*height+2*width*height,"inches squared.")
```
