#Intro to Machine Learning (Day 3)

```python
import pandas as pd

#pd now stands for pandas (can be anything you please)
```

```python
#saves file path as a variable
melb_file_path = '/content/melb_data.csv.zip'

#read the data and store it in Dataframe
melb_data = pd.read_csv(melb_file_path)

#print a summary of the data in melb_data.csv
melb_data.describe()

```

```python
melb_data.columns
#shows columns of the data
```

```python
melb_data = melb_data.dropna(axis=0)
#dropna drops missing values (drop N/A)
```

There are different ways of selecting subsets of your data. We will focus on two.
1. Dot notation, which we use to select the "prediction target"
2. Selecting with a column list, which we use to select a specific feature(s)

#Selecting the Target Prediction

```python
y = melb_data.Price
```

#Choosing "Features"

```python
melb_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
X = melb_data[melb_features]
```

```python
X.describe()
```

```python
X.head()
```

#Building Our Model
Using Scikit-learn library (sklearn)

Steps to building a model:
1. Define: What type of model will it be? A decision tree, or some other type of model?
2. Fit: Capture patterns from the data provided.
3. Predict
4. Evaluate


```python
from sklearn.tree import DecisionTreeRegressor

#Defined the model
melb_model = DecisionTreeRegressor(random_state=1)

#Fit the model
melb_model.fit(X, y)
```

#comment



```python
print("We will make predictions for the following 5 houses:")
print(X.head())
print("The predictions are ")
print(melb_model.predict(X.head()))
```

```python
melb_data.head()
```

#Model Validation



*   MAE: Mean Absolute Error

Error = (actual value) - (predicted value)



```python
from sklearn.metrics import mean_absolute_error
#Import the mean absolute error function from sklearn.metrics

predicted_home_prices = melb_model.predict(X)
mean_absolute_error(y, predicted_home_prices)

```

#Split

```python
from sklearn.model_selection import train_test_split

```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# We split the data into the training part and the validation/evaluation part.
# We do this for both the target and the features.
# Since we use random_state = 0, this means that we don't jumble the data when we split it.

melb_model2 = DecisionTreeRegressor()
melb_model2.fit(train_X, train_y)

#Get prediction now

val_predictions = melb_model2.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```
